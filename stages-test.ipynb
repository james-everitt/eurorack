```python
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np
from glob import glob
```

```python
new = {}
stock = {}
for fn in glob("stages_*wav"):
    n = wavfile.read(fn)[1]
    new[fn[:-4]] = n
    try:
        s = wavfile.read(f"stages_test_factory/{fn}")[1]
        stock[fn[:-4]] = s
    except:
        print(f"stock file not found for {fn}; skipping")
        continue
    if (s==n).all():
        print(f"✅ {fn}")
    else:
        print(f"❌ {fn}")
        for i, label in enumerate(["gate", "value", "segment", "phase"]):
            if (s.T[i] == n.T[i]).all():
                print(f"\t✅ {label}")
            else:
                print(f"\t❌ {label}")
                #plt.figure(figsize=(16,4))
                #plt.plot(s.T[i])
                #plt.plot(n.T[i])
                #plt.title(f"{fn} - {label}")
```
